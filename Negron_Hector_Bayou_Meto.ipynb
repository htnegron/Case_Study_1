```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt

# reading the data into a pandas df
df = pd.read_csv('Data/BayouMeto_Lidar_QA.csv')

# plot the x and y columns (locations of checkpoints)
df.plot(x = 'X', y = 'Y', kind = 'scatter')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Locations of Checkpoints')
plt.show()
```

```python
# Compute and print the mean and standard deviation of the differences between the Lidar-Z and GPS-Z

# calculate and print the mean of the difference of Z values
DiffZ = df['Lidar-Z'] - df['GPS-Z']  # this does the difference for each value in the column, not the entire column
MeanDiffZ = DiffZ.mean()
print('The mean of the difference between Lidar-Z and GPS-Z is ' + str(MeanDiffZ))
print('\n')
#calculate and print the stdev of the differences
StdZ = DiffZ.std()
print('The standard deviation of the differences between the Lidar-Z and GPS-Z is ' + str(StdZ))
```

```python
# Compute and print the mean and standard deviation of the differences between the LiDAR-Z and GPS-Z measurements by Region.

# Separate the dataframe into regions
North = df[df['Region'] == 'North']
South = df[df['Region'] == 'South']

# calculate the difference between Lidar and GPS for the North
NDiffZ = North['Lidar-Z'] - North['GPS-Z']

# calculate the mean of the difference for the North
MeanNDiff = NDiffZ.mean()
print('The mean of the difference between Lidar and GPS Z values in the North is: ' + str(MeanNDiff))
print('\n')
# calculate the difference between Lidar and GPS for the South
SDiffZ = South['Lidar-Z'] - South['GPS-Z']

# calculate the mean of the difference
MeanSDiff = SDiffZ.mean()
print('The mean of the difference between Lidar and GPS Z values in the South is: ' + str(MeanSDiff))
print('\n')
# calculate and print the stdev of the North
NStd = NDiffZ.std()
print('The standard deviation of the difference between Lidar-Z and GPS-Z for the North is: ' + str(NStd))
print('\n')
# calculate and print the stdev of the South
SStd = SDiffZ.std()
print('The standard deviation of the difference between Lidar-Z and GPS-Z for the South is: ' + str(SStd))


```

```python
# Compute and print the mean and standard deviation of the differences betweent the LiDAR-Z and GPS-Z measurements by LandCover.

# Separate the dataframe into landcovers
Asphalt = df[df['LandCover'] == 'Asphalt']
Dirt = df[df['LandCover'] == 'Dirt']
Forest = df[df['LandCover'] == 'Forest']
Grass = df[df['LandCover'] == 'Grass']
Scrub = df[df['LandCover'] == 'Scrub']

# calculate the difference
ADiff = Asphalt['Lidar-Z'] - Asphalt['GPS-Z']
# calculate and print the mean of the difference
MeanADiff = ADiff.mean()
print('The mean of the difference between Lidar-Z and GPS-Z for Asphalt is: ' + str(MeanADiff))

# calculate and print the stdev of the difference
StdADiff = ADiff.std()
print('The standard deiation of the difference between Lidar-Z and GPS-Z for Asphalt is: ' + str(StdADiff))
print('\n')
# calculate the difference
DDiff = Dirt['Lidar-Z'] - Dirt['GPS-Z']
# calculate and print the mean of the difference
MeanDDiff = DDiff.mean()
print('The mean of the difference between Lidar-Z and GPS-Z for Dirt is: ' + str(MeanDDiff))

# calculate and print the stdev of the difference
StdDDiff = DDiff.std()
print('The standard deiation of the difference between Lidar-Z and GPS-Z for Dirt is: ' + str(StdDDiff))
print('\n')
# calculate the difference
FDiff = Forest['Lidar-Z'] - Forest['GPS-Z']
# calculate and print the mean of the difference
MeanFDiff = FDiff.mean()
print('The mean of the difference between Lidar-Z and GPS-Z for Forest is: ' + str(MeanFDiff))

# calculate and print the stdev of the difference
StdFDiff = FDiff.std()
print('The standard deiation of the difference between Lidar-Z and GPS-Z for Forest is: ' + str(StdFDiff))
print('\n')
# calculate the difference
GDiff = Grass['Lidar-Z'] - Grass['GPS-Z']
# calculate and print the mean of the difference
MeanGDiff = GDiff.mean()
print('The mean of the difference between Lidar-Z and GPS-Z for Grass is: ' + str(MeanGDiff))

# calculate and print the stdev of the difference
StdGDiff = GDiff.std()
print('The standard deiation of the difference between Lidar-Z and GPS-Z for Grass is: ' + str(StdGDiff))
print('\n')
# calculate the difference
SDiff = Scrub['Lidar-Z'] - Scrub['GPS-Z']
# calculate and print the mean of the difference
MeanSDiff = SDiff.mean()
print('The mean of the difference between Lidar-Z and GPS-Z for Scrub is: ' + str(MeanSDiff))

# calculate and print the stdev of the difference
StdSDiff = SDiff.std()
print('The standard deiation of the difference between Lidar-Z and GPS-Z for Scrub is: ' + str(StdSDiff))
```

The average of the differences in Lidar vs. GPS values is much higher in the North than the South. The standard deviation of the differences between Lidar vs. GPS values is slightly higher in the North than the South. 

The highest average difference in values based on land cover is Forest. The lowest average difference in values based on land cover is dirt. 

The highest standard deviation in values based on land cover is also forest. The lowest standard deviation in values based on land cover is dirt. 

Land covered by forest has the biggest differences between Lidar and GPS values, and land covered by dirt has the smallest differences between the two methods. 

It is likely that the North region has more checkpoints that have land covered by forest, scrub, and grass since they tend to have a higher mean difference and stdev.
